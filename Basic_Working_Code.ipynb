```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
#defining x and y points
x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21,21]
```

```python
#creating the data points
data = list(zip(x, y))
#visualising the data points
print(data)
```

```python
#defining the model
hierarchical_cluster = AgglomerativeClustering(n_clusters=2,metric='euclidean', linkage='single')
#gettings clusters
labels = hierarchical_cluster.fit_predict(data)
```

```python
#plotting the clusters thus obtained
plt.scatter(x, y, c=labels)
plt.show()
```

```python
#importing dendrograms
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#getting the distances between clusters
linkage_data = linkage(data, method='single', metric='euclidean')
```

```python
#defining the dendrogram
dendrogram(linkage_data)
#plotting the graph
plt.show()
```
